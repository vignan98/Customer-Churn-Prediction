```python
def print_unique_col_values(df):
    for column in df:
            if df[column].dtype=='object':
                print(f'{column}:{df[column].unique()}')
```

```python
print_unique_col_values(df1)
```

```python
df1.replace('No internet service','No',inplace=True)
df1.replace('No phone service','No',inplace=True)
```

```python
print_unique_col_values(df1)
```

```python
df1.replace({'Yes':1,'No':0},inplace=True)
```

```python
for column in df1:
    print(f'{column}:{df1[column].unique()}')
```

```python
df1['gender'].replace({'Male':0,'Female':1},inplace=True)
```

```python
for column in df1:
    print(f'{column}:{df1[column].unique()}')
```

```python
df2=pd.get_dummies(data=df1,columns=['InternetService','Contract','PaymentMethod'])
df2.columns
```

```python
df2.sample(100)
```

```python
df2.dtypes
```

```python
col_to_scale=['MonthlyCharges','TotalCharges','tenure']
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
df2[col_to_scale]= scaler.fit_transform(df2[col_to_scale])
```

```python
X=df2.drop('Churn',axis='columns')
y=df2['Churn']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=5)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
#20,15,1 represents number of output neurons in each layer
#input_shape(26,) represents input layer neurons and it is 26 neurons in first layer input which is equal to number of features
#hidden layers do not require input neurons count


import tensorflow as tf
from tensorflow import keras
model=keras.Sequential([
                keras.layers.Dense(20, input_shape=(26,),activation='relu'),
                keras.layers.Dense(15,activation='relu'),
                keras.layers.Dense(1,activation='sigmoid'),


    
])

model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
model.fit(X_train,y_train,epochs=100)
```

```python
model.evaluate(X_test,y_test)
```

```python
y_pred=model.predict(X_test)
```

```python
y_pred[:5]
```

```python
y_pred1=[]
for i in y_pred:
    if i>0.5:
        y_pred1.append(1)
    else:
        y_pred1.append(0)
```

```python
y_pred1[:10]
```

```python
y_test[:10]
```

```python
from sklearn.metrics import confusion_matrix,classification_report
print(classification_report(y_test,y_pred1))
```

```python
import seaborn as sns
cm=tf.math.confusion_matrix(labels=y_test,predictions=y_pred1)
plt.figure(figsize=(10,7))
sns.heatmap(cm,annot=True,fmt='d')
plt.xlabel('predicted')
plt.ylabel('Truth')
```
